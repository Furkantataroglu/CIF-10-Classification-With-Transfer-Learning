```python
from tensorflow.keras.datasets import cifar10
import tensorflow  as tf
import numpy as np
import pandas as pd
import os
from PIL import Image
from tensorflow.keras import layers, models, callbacks
from tensorflow import keras
from tensorflow.keras.applications import ResNet50V2

import matplotlib.pyplot as plt
```

```python
# 1) CIFAR-10’u indir ve yükle  ----------------------------
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python

# 2) Boyutları kontrol et ---------------------------------
print(f"Train images : {x_train.shape}")   # (50000, 32, 32, 3)
print(f"Train labels : {y_train.shape}")   # (50000, 1)
print(f"Test  images : {x_test.shape}")    # (10000, 32, 32, 3)
print(f"Test  labels : {y_test.shape}")    # (10000, 1)
```

# CONSTANTS (hyper-parameters & config)

```python
NUM_CLASSES   = 10                    # CIFAR-10 → 10 etiket
INPUT_SHAPE   = (32, 32, 3)           # H x W x C #from cfr10 site.
IMG_SIZE = 224 # RESNET 224 X 224 X 3 IMAGE BEKLER
CLASS_NAMES   = [
    "airplane", "automobile", "bird", "cat", "deer",
    "dog", "frog", "horse", "ship", "truck"
    
]
EPOCHS=30
BATCH_SIZE = 64
VAL_RATIO    = 0.1                # 10 % validation
```

# train validation test split and preprocess

```python
# Etiketleri (N,1) → (N,) düzleştir; NumPy uyarısını önler
y_train = y_train.flatten()
y_test  = y_test.flatten()
```

```python
# ------------------------------------------------
# 3) Train → (train + val) ayır
# ------------------------------------------------
num_val          = int(len(y_train) * VAL_RATIO)          # 5 000
x_val, y_val     = x_train[-num_val:], y_train[-num_val:]
x_train, y_train = x_train[:-num_val], y_train[:-num_val]
```

```python

# Datasetleri oluştur
train_ds = tf.data.Dataset.from_tensor_slices((x_train, y_train)).shuffle(10000).batch(BATCH_SIZE)
val_ds = tf.data.Dataset.from_tensor_slices((x_val, y_val)).batch(BATCH_SIZE)
test_ds = tf.data.Dataset.from_tensor_slices((x_test, y_test)).batch(BATCH_SIZE)



data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal_and_vertical"),  # hem yatay hem dikey
    layers.RandomRotation(0.2),                    # ±20 % → ~±36°
])

resize_and_rescale = tf.keras.Sequential([
    layers.Resizing(IMG_SIZE, IMG_SIZE),
    layers.Rescaling(1./255),
])

# 1) Eğitim için preprocess   (augmentation + normalize)
def preprocess_train(img, label):
    img = data_augmentation(img, training=True)  # sadece eğitimde aktif
    img = resize_and_rescale(img)
    return img, label

# 2) Doğrulama/Test için preprocess   (yalnızca normalize)
def preprocess_eval(img, label):
    img = resize_and_rescale(img)
    return img, label                      # one-hot gerekiyorsa burada yap



train_ds = train_ds.map(
    preprocess_train,            # augmentation + normalize
    num_parallel_calls=tf.data.AUTOTUNE
).prefetch(tf.data.AUTOTUNE)

val_ds = val_ds.map(
    preprocess_eval,             # sadece normalize
    num_parallel_calls=tf.data.AUTOTUNE
).prefetch(tf.data.AUTOTUNE)

test_ds = test_ds.map(
    preprocess_eval,             # sadece normalize
    num_parallel_calls=tf.data.AUTOTUNE
).prefetch(tf.data.AUTOTUNE)
```

```python

print(f"Train images : {len(x_train)}")     
print(f"Val   images : {len(x_val)}")       
print(f"Test  images : {len(x_test)}") 
```

## samples

```python
# Pick 9 samples (first 9 for simplicity)
num_samples = 9
samples = x_train[:num_samples]
labels  = y_train[:num_samples]

# Plot
plt.figure(figsize=(6, 6))
for i, (img, label) in enumerate(zip(samples, labels)):
    plt.subplot(3, 3, i + 1)
    plt.imshow(img)
    plt.title(CLASS_NAMES[label], fontsize=8)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

# RESNET TRANSFER LEARNING

```python
# ------------------------------------------------------------------
#  Build the ResNet50V2 transfer-learning model
# ------------------------------------------------------------------
base_model = ResNet50V2(
    include_top=False,          # drop the 1000-class FC head
    weights="imagenet",         # load ImageNet weights
    input_shape=(IMG_SIZE, IMG_SIZE, 3),
    pooling="max",              # add GlobalMaxPooling2D automatically
    name="resnet50v2",
)
base_model.trainable = False    # freeze backbone for the first stage

# RESNETİ KURDUK ŞİMDİ KENDİ MODELİMİZE EKLİYORUZ.
model = models.Sequential([
    base_model,
    layers.Dense(256, activation="relu"),    # ara katman
    layers.Dropout(0.3),
    layers.Dense(NUM_CLASSES, activation="softmax", name="pred")
], name="resnet50v2_cifar10")
```

```python

model.compile(
    optimizer=keras.optimizers.Adam(1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
)

history_feat = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10,
    callbacks=[
        keras.callbacks.EarlyStopping(
            monitor="val_accuracy", patience=3, restore_best_weights=True
        )
    ],
)
```
